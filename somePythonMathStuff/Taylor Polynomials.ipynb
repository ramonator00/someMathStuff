# Taylor Polynomials

\begin{align}
\displaystyle \underbrace{f(x) = f(x_0) + \frac{f'(x_0)}{1!}(x - x_0)^{1} + \frac{f''(x_0)}{2!}(x - x_0)^{2} + ... + \frac{f^{n}(x_0)}{n!}(x - x_0)^{n}}_{\substack{Taylorisches\,Polynom\,f_n(x)\,vom\,Grade\,n}} + \underbrace{R_n(x)}_{\substack{Restglied}}
\end{align}

\begin{align}
\displaystyle R_n(x) = \frac{f^{(n + 1)}(\xi)}{(n + 1)!}(x - x_0)^{n + 1}
\end{align}

```python
import sympy as sy
from sympy.functions import sin, cos, ln, atan2
import matplotlib.pyplot as plt

def fakultaet(n):
    """
    Diese Funktion berechnet die Fakultät
    :param n: Die zu berechnenden Zahl n
    :return: Fakultät von n
    """
    if n <= 0:
        return 1
    else:
        return n * fakultaet(n-1)
```

```python
def taylor(f, x0, n):
    """
    Approximation einer Funktion durch das Taylorpolynom
    :param f: Die zu berechnende Funktion
    :param x0: Die zu berechnende Stelle der Funktion
    :param n: Die n-te Ordnung der Funktion
    :return: 
    """
    i = 0
    p = 0
    while i <= n:
        p = p + (f.diff(x, i).subs(x, x0))/(fakultaet(i))*(x-x0)**i
        i += 1
    return p

x = sy.Symbol('x')
f = sin(x)
print("Die zu berechnende Funktion für sin(x) lautet:")
taylor(f, 0, 3)
```

```python
f1 = ln(x + 1)
print("Die zu berechnende Funktion für ln(x + 1) lautet:")
taylor(f1, 0, 3)
```

```python
f2 = atan2(x, 1)
print("Die zu berechnende Funktion für arctan(x) lautet:")
taylor(f2, 0, 3)
```

```python
import numpy as np

def plot(f):
    """
    Die Funktion visualisiert die Taylorpolynome in einer Grafik
    :param f: Die zu berechnende Funktion
    :return: Polynome n-ten Grade als Grafik
    """
    plt.style.use('ggplot')
    fig = plt.figure(figsize=(8, 5))
    x_lims = [-5, 5]
    x1 = np.linspace(x_lims[0],x_lims[1], 200)
    y1 = []
    
    for j in range(1, 10, 2):
        func = taylor(f, 0, j)
        for k in x1:
            y1.append(func.subs(x, k))
        plt.plot(x1, y1, label='Polynom '+ str(j) + '. Grades')
        y1 = []
        
    plt.title('Taylorpolynom {0}'.format(f))
    plt.xlim(x_lims)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(loc='best')
    plt.show()

f = sin(x)
plot(f)

f1 = ln(x + 1)
plot(f1)

f2 = atan2(x, 1)
plot(f2)
```
